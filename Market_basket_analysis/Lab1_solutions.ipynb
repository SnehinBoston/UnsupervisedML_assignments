<b> Q2. What relationship between these variables keeps the number of interesting association rules in this range? <b>

- If we have very low support, then we need to increase the confidence to more than 10 times the support. In this situation, slight changes in lift and confidence have a strong impact on the number of rules.(support = 0.010, confidence = 0.14, lift = 1.2).

- If we have high support values, then even huge decrease in confidence and lift values do not have any effect on the number of rules.

```python
# Support and confidence
def association_function(support, confidence, lift):
    association_rules = apriori(lab_mba_list, min_support = support, min_confidence = confidence
                        ,min_lift = lift)
    association_results = list(association_rules)
    return len(association_results)
```

```python
# Support: 0.018, confidence: 0.08, lift: 1 - 10000000
length = association_function(support = 0.010, confidence = 0.14, lift = 1.2)
print(length)
```

```python
# Sort the values by support
rules_df = pd.DataFrame(columns = ('Items','Support'))
Support = []
Confidence = []
Lift = []
Items = []
Antecedent = []
Consequent = []

for RelationRecord in sorted_75:
    Support.append(RelationRecord.support)
    Items.append(RelationRecord.items)
```
